### Import quotes_fetcher

```python
from quotes_fetcher import Symbols
```

### Fetch data

Let's get data for Apple, Google and Tesla for the last 10 days.

```python
symbols = Symbols('aapl goog tsla', period='10d')
symbols.quotes
```

We can also specify a more exact period via start and end time thresholds.

```python
symbols = Symbols('aapl goog tsla', period_start='2018-04-26', period_end='2019-09-13')
symbols.quotes
```

### Get more data

If we don't like the default view, we can easily transpose the table. Let's do it.

```python
symbols.quotes_t
```

Can we get more data? For sure. Let's check what we've got. 

```python
symbols = Symbols('yndx, msft ibm , KO tsla aApl BAC ', period='5y')
symbols.metrics
```

By the way, if we need to get the list of symbols data we've obtained for, it's not a big deal. 

```python
symbols.symbols
```
